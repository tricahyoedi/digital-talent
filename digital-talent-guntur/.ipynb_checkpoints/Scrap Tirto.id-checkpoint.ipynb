```python
import requests

page = requests.get("https://tirto.id/indeks/2")
html_page = page.content
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(html_page, "lxml")
print(soup.prettify())
```

```python
import json,csv

links = soup.find_all('script')
isi_script = links[4].text.replace('window.__NUXT__=','')[:-1]
article = json.loads(isi_script)
with open('tirto_links.csv', 'w') as csvfile:
    writer = csv.writer(csvfile)
    for a in article['data'][0]['listarticle']:
        writer.writerows([str(a['articleUrl']),str(a['date_news'])])
```
