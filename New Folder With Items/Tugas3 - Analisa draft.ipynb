# 1. Deskripsi Data

Data didapat dari https://data.go.id/dataset/realisasi-sumber-pendapatan-pemerintah-apbn dengan judul "Realisasi Sumber Pendapatan Pemerintah APBN" yang berisi mengenai realisasi pos pendapatan Pemerintah di APBN dari tahun 1990 sampai dengan 2013.

Data terdiri dari 4 kolom yaitu
- tahun: Tahun APBN
- sumber_pendapatan: Realisasi Sumber Pendapatan Pemerintah (Pajak / Bukan Pajak / Hibah)
- keterangan_sumber_pendapatan: Keterangan dari Realisasi Sumber Pendapatan tersebut
- nilai: Nilai dari Realisasi Pos Pendapatan tersebut (dalam Milyar Rupiah)

Berikut adalah sample data:
